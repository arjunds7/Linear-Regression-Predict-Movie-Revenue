```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(12,6))
plt.scatter(X,y, alpha=0.4)
plt.title("Film cost vs Global revenue")
plt.xlabel('Production budget')
plt.ylabel('Gross revenue')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)

plt.show()
```

```python
regression = LinearRegression()
regression.fit(X,y)
```

```python
#Finding the slope coefficient _theta_1
```

```python
regression.coef_
```

```python
#Finding the intercept _theta_0
    
```

```python
regression.intercept_
```

```python
#plt.plot(X, regression.predict(X) .. plotting X value(budget) against predicted y(predicted gross revenue),  regression.predict(X)=y
```

```python
plt.figure(figsize=(12,6))
plt.scatter(X,y, alpha=0.4)
plt.plot(X, regression.predict(X), color='red', linewidth=4)
plt.title("Film cost vs Global revenue")
plt.xlabel('Production budget')
plt.ylabel('Gross revenue')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
# Finding how good our prediction is which is 55%
```

```python
regression.score(X,y)
```
